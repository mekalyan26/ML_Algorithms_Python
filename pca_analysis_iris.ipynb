```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd
import numpy as np
```

```python
data = load_iris()
```

```python
data.DESCR.split('\n')
```

```python
features = data.data
target = data.target
features.shape, target.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(features, target, train_size = 0.8, random_state = 3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
my_model = LogisticRegression()
```

```python
my_model.fit(X_train,y_train)
```

```python
preds = my_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(accuracy_score(y_test,preds))
```

```python
print(confusion_matrix(y_test,preds))
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X_train)
```

```python
X_train_transformed = pca.transform(X_train)
```

```python
pca.explained_variance_ratio_
```

```python
X_test_transformed = pca.transform(X_test)
```

```python
my_model.fit(X_train_transformed, y_train)
```

```python
preds_pca = my_model.predict(X_test_transformed)
```

```python
print(accuracy_score(y_test, preds_pca))
```

```python
print(confusion_matrix(y_test, preds_pca))
```
