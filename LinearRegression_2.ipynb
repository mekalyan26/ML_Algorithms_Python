
### LASSO Regression - Lasso regression stands for Least Absolute Shrinkage and Selection Operator (L1 - Regularization)
###                  - Adds penalty term to the cost function (absolute sum of coefficients )

### Ridge Regression - Adds penalty term square of the coefficient (L2 - Regularization)
###                  - Adds coefficient lambda to control the penalty term 

### Elastic Net - Combines regularization of both lasso and Ridge. (Combines L1 and L2 regularizations)
